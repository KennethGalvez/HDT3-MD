```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('wine_fraud.csv')
numeric_data = data.select_dtypes(include='number')

# Crear una figura y subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))

# Histogramas
numeric_data.hist(ax=axs[0, 1])
axs[0, 1].set_title('Histogramas')

# Mostrar
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('wine_fraud.csv')
numeric_data = data.select_dtypes(include='number')

# Diagramas de caja y bigotes
fig, axs = plt.subplots(figsize=(10, 6))
numeric_data.plot.box(ax=axs)
axs.set_title('Diagramas de caja y bigotes')
axs.set_xlabel('Variables')
axs.set_ylabel('Valor')
axs.set_xticklabels(axs.get_xticklabels(), rotation=45, ha='right')

# Mostrar las gráficas
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('wine_fraud.csv')
numeric_data = data[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides']] # seleccionar solo las columnas de interés

# Gráfico de dispersión
fig, axs = plt.subplots(figsize=(10, 10))
pd.plotting.scatter_matrix(numeric_data, diagonal='hist', ax=axs)
axs.set_title('Gráfico de dispersión')
axs.set_xlabel('Variables')
axs.set_ylabel('Valor')

# Mostrar las gráficas
plt.show()
```
